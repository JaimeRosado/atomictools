```python
import atomictools as at
```

## Radial functions

```python
R_32 = at.R_hydrog(n=3, l=2)
R_32.plot_P2()
```

```python
R_positronium_32 = at.R_hydrog(n=3, l=2, mu=0.5)
R_positronium_32.plot_P2()
```

```python
R_FeVII_2s = at.R_num("FeVII_2s.txt")
R_FeVII_2s.plot_P()
```

## Angular functions

```python
Y_21 = at.spherical_harmonic(l=2, m=1)
Y_21.plot_prob()
```

```python
Y_21.plot_phase()
```

```python
d_yz = at.real_ang_function(l=2, m=1, part="Im")
d_yz.plot_phase()
```

```python
d_yz2 = at.comb_ang_function(functions=[(2, 1), (2, -1)], coefficients=[1j, -1j])
d_yz.plot_phase()
```
